## Situation 1: 10am

### Index

```python
import pandas as pd
```

```python
df_index = pd.read_csv('每小時指數.csv')
df_index
```

```python
df_index['date'] = df_index['date'].astype(str).str.strip("'")
df_index['date'] = pd.to_datetime(df_index['date'], format="%Y/%m/%d %H:%M")
df_filtered = df_index[df_index['date'].dt.hour.isin([10, 14])]
df_filtered
```

```python
# Extracting just the date part from the datetime
df_filtered['date_only'] = df_filtered['date'].dt.date

# Grouping by the date
grouped = df_filtered.groupby('date_only')

index_data = grouped.agg(index_9=('open', 'first'), 
                               index_10=('close', 'first'), 
                               index_2=('close', 'last')).reset_index()
index_data
```

### Future

```python
df_future = pd.read_csv('指數期貨.csv')
df_future
```

```python
df_future['date'] = df_future['date'].astype(str).str.strip("'")
df_future['datetime'] = pd.to_datetime(df_future['date'].astype(str).str.strip("'") + ' ' + df_future['time'], format='%Y/%m/%d %H:%M')
df_filtered = df_future[df_future['datetime'].dt.strftime('%H:%M').isin(['09:45', '13:45'])]

df_filtered = df_filtered.drop(['date', 'time'], axis=1)
df_filtered
```

```python
# Extracting just the date part from the datetime
df_filtered['date_only'] = df_filtered['datetime'].dt.date

# Grouping by the date
grouped = df_filtered.groupby('date_only')

future_data = grouped.agg(future_9=('open', 'first'), 
                               future_10=('close', 'first'), 
                               future_2=('close', 'last')).reset_index()
future_data
```

### Merge

```python
merged_df = pd.merge(future_data, index_data, on='date_only')
merged_df
```

```python
merge_df = merged_df.sort_index(axis=1)
merge_df.to_csv('10am_rowdata.csv', index=False)
```

## Situation 2: tomorrow

```python
import pandas as pd
```

```python
def merge_csv_on_date(filepaths):
    dfs = []
    
    for filepath in filepaths:
        print(filepath)
        df = pd.read_csv(filepath)
        df['date'] = pd.to_datetime(df['date'])
        df['date'] = df['date'].dt.strftime('%Y/%m/%d')
        dfs.append(df)
    
    merged_df = pd.concat(dfs, join='outer', sort=True).groupby('date').sum().reset_index()
    return merged_df
```

```python
filepaths = ['日圓.csv', '美元台幣.csv', 'TaiwanFuturesExchange.csv', '大盤日開高收低.csv', '2330ADR.csv', '費半.csv']
print(filepaths)
```

```python
merge_df = merge_csv_on_date(filepaths)
```

```python
merge_df.replace(0.0, -1, inplace=True)
```

```python
rows_to_keep = (merge_df == -1).sum(axis=1) < 5
merge_df = merge_df[rows_to_keep]
merge_df
```

```python
merge_df.columns
```

```python
dropped_columns = ['台積_SMA10', '台積_SMA5', '大盤_SMA10', '大盤_SMA5', '日圓_SMA10', '日圓_SMA5', '費半_SMA10', '費半_SMA5']
merge_df = merge_df.drop(columns=dropped_columns)
merge_df = merge_df.sort_index(axis=1)
```

```python
merge_df.to_csv('row_data.csv', index=False)
```
