```python
import pandas as pd
```

```python
def merge_csv_on_date(filepaths):
    dfs = []
    
    for filepath in filepaths:
        print(filepath)
        df = pd.read_csv(filepath)
        df['date'] = pd.to_datetime(df['date'])
        df['date'] = df['date'].dt.strftime('%Y/%m/%d')
        dfs.append(df)
    
    merged_df = pd.concat(dfs, join='outer', sort=True).groupby('date').sum().reset_index()
    return merged_df
```

```python
filepaths = ['日圓.csv', '美元台幣.csv', 'TaiwanFuturesExchange.csv', '大盤日開高收低.csv', '2330ADR.csv', '費半.csv']
print(filepaths)
```

```python
merge_df = merge_csv_on_date(filepaths)
```

```python
merge_df.replace(0.0, -1, inplace=True)
```

```python
rows_to_keep = (merge_df == -1).sum(axis=1) < 5
merge_df = merge_df[rows_to_keep]
merge_df
```

```python
dropped_columns = ['台積_SMA10', '台積_SMA5', '大盤_SMA10', '大盤_SMA5', '日圓_SMA10', '日圓_SMA5', '費半_SMA10', '費半_SMA5']
merge_df = merge_df.drop(columns=dropped_columns)
```
